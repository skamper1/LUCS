```python
# 01_starter_plotting.py
# ----------------------
# Goal: Load the wildlife dataset and create your first line chart.

import pandas as pd
import matplotlib.pyplot as plt

# 1) Load data
df = pd.read_csv('wildlife_reserve.csv')
print(df.head())  # Peek at the data

# 2) Create a line chart
plt.figure(figsize=(8,4))
plt.plot(df['Year'], df['Rabbits'], marker='o')  # TODO: choose a color, e.g. color='green'

# 3) Labels and title
plt.title('Rabbit Population Over Time')  # TODO: edit the title text if you like
plt.xlabel('Year')
plt.ylabel('Number of Rabbits')

# 4) Optional styling
plt.grid(True, alpha=0.3)

# 5) Display the chart
plt.show()
```
