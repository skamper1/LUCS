```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('wildlife_reserve.csv')

fig, ax1 = plt.subplots(figsize=(9,4))

# Left axis: populations
ax1.plot(df['Year'], df['Rabbits'], color='green', marker='o', label='Rabbits')
ax1.plot(df['Year'], df['Foxes'], color='orange', marker='s', label='Foxes')
ax1.set_xlabel('Year')
ax1.set_ylabel('Population', color='black')
ax1.grid(True, alpha=0.3)
ax1.tick_params(axis='y', labelcolor='black')

# Right axis: temperature
ax2 = ax1.twinx()
ax2.plot(df['Year'], df['AvgTemperature'], color='crimson', marker='^', linewidth=2, label='Avg Temp (°C)')

# Legend: combine handles from both axes
lines_1, labels_1 = ax1.get_legend_handles_labels()
lines_2, labels_2 = ax2.get_legend_handles_labels()
ax1.legend(lines_1 + lines_2, labels_1 + labels_2, loc='best')

plt.title('Rabbits & Foxes vs Average Temperature')
plt.tight_layout()
plt.savefig('plot_data.png', dpi=150)
plt.show()
```
